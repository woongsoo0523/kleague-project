```python
import pandas as pd
import numpy as np

df = pd.read_csv('./test.csv',encoding='euc-kr')

df["home"]=df["점유율"].str[0:5]
df["away"]=df["점유율"].str[6:]
d1 = df[["home","away"]]
df['sample'] = df[['홈','어웨이']].iloc[0][0]+df[['홈','어웨이']].iloc[0][1]

dft = df[["sample","home","away"]]


for i in range(len(dft)):
    dft['sample'].loc[i] = df[['홈','어웨이']].iloc[i][0]+df[['홈','어웨이']].iloc[i][1]

dft1 = dft.set_index("sample").head(10)
dft1.astype(float).plot.bar()
```
